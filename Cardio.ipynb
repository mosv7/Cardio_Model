```python
import pandas as pd  
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('cardio_train.csv',sep = ';')
df.head()
```

```python
df= df.drop('id' , axis = 'columns')
df.head()
```

```python
df.head()
```

```python
df['age'] = np.round(df['age'] / 365)
df.head()
```

```python
df.head()
```

```python
df['gender'].replace([1,2],value = ['Female','Male'],inplace = True)
df.head()
```

```python
df.head()
```

```python
df[['smoke','alco','active','cardio']] = df[['smoke','alco','active','cardio']].replace([0,1],value = ['no','yes'])
df.head()
```

```python
df[['cholesterol','gluc']] = df[['cholesterol','gluc']].replace({1: 'normal', 2: 'above normal', 3: 'well above normal'})
df.head()
```

```python
df.drop(df[(df['ap_hi'] > 240)|(df['ap_hi'] <40) | (df['ap_lo'] >240) |  (df['ap_lo'] < 40)].index, inplace=True)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['ap_hi'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['ap_lo'])
plt.show()
```

```python
df['BMI'] = df['weight'] / (df['height']/100)**2
df.head()
```

```python
conditions = [
    ((df['ap_hi'] < 90) | (df['ap_lo'] < 60)),  # Low blood pressure
    ((df['ap_hi'] >= 90) & (df['ap_hi'] < 120) & (df['ap_lo'] >= 60) & (df['ap_lo'] < 80)),  # Normal
    (((df['ap_hi'] >= 120) & (df['ap_hi'] < 140)) | ((df['ap_lo'] >= 80) & (df['ap_lo'] < 90))),  # Lightly high
    ((df['ap_hi'] >= 140) | (df['ap_lo'] >= 90))  # High
]

# Define the corresponding choices
choices = ['low', 'normal', 'lightly high', 'high']

# Use np.select() to create the new column, with a default value for unmatched rows
df['blood_pressure'] = np.select(conditions, choices, default='unknown')

# Display the first few rows of the DataFrame
df.head()
```

```python
df['blood_pressure'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder,LabelEncoder

le = LabelEncoder()
```

```python
cols = ['cholesterol', 'gluc', 'smoke', 'alco', 'active','cardio', 'gender']
for col in cols:
    df[col+'_enc'] = le.fit_transform(df[col])

df.head()
```

```python
bb = df.groupby(['blood_pressure' , 'cardio'])[['cardio_enc']].count().reset_index().rename(columns={'cardio_enc':'count'})
bb
```

```python
X = df[['age', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol_enc', 'gluc_enc', 'smoke_enc', 'alco_enc', 'active_enc'] ]
y = df['cardio_enc']   
```

```python
X.head()
```

```python
y.head()
```

```python
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size=0.24, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score , precision_score, recall_score, f1_score , roc_auc_score, confusion_matrix
```

```python
Cls = RandomForestClassifier(n_estimators=100, criterion='gini', max_depth=5, min_samples_split=4, min_samples_leaf=2, min_weight_fraction_leaf=0.0, max_features='sqrt', max_leaf_nodes=None, min_impurity_decrease=0.0, bootstrap=True, oob_score=False, n_jobs=None, random_state=None, verbose=0, warm_start=False, class_weight=None, ccp_alpha=0.0, max_samples=None, monotonic_cst=None)
Cls.fit(X_train, y_train)

y_pred = Cls.predict(X_test)
print("Accuracy:", Cls.score(X_test, y_test))
print("Precision:", precision_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC AUC Score:", roc_auc_score(y_test, y_pred))
```

```python
cn = confusion_matrix(y_test, y_pred)
cn
```
